```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily

# Set time period
start = datetime(2012, 1, 1)
end = datetime(2021, 12, 31)


# Create Point for Vancouver, BC
london= Point(51.489096, -0.191170, 45)

# Get daily data for 2018
data = Daily(london, start, end)
data = data.fetch()
time = pd.DataFrame(data, columns=['tavg', 'prcp', 'snow', 'wspd'])
time['snow'] = time['snow'].fillna(0)
time['prcp'].fillna(method='ffill', inplace=True)
time['tavg'].fillna(method='ffill', inplace=True)
time['wspd'].fillna(method='ffill', inplace=True)
time['time'] = time.index


# Plot line chart including average, minimum and maximum temperature
data.plot(y=["prcp"])
plt.show()
```
